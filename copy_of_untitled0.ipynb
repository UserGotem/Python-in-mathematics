```python
import matplotlib.pyplot as plt
import numpy as np

def scatter_plot_game():
  plt.figure(figsize=(8, 8))
  x = np.random.randint(-20, 20, 15)
  y = np.random.randint(-20, 20, 15)

  plt.scatter(x, y, marker='o', color='blue')
  plt.title('Scatter Plot Game')
  plt.xlabel('X-axis')
  plt.ylabel('Y-axis')

  player_x = float(input("Enter X-coordinate: "))
  player_y = float(input("Enter Y-coordinate: "))

  if (player_x in x) and (player_y in y):
      print("That's right!")
  else:
    print("Not right. Try again it.")

  plt.show()

scatter_plot_game()
```

```python
import random

def algebra_practice_game():
  a = random.randint(-15, 15)
  b = random.randint(-15, 15)
  c = random.randint(-30, 30)

  equation = f"{a}x+{b}={c}"

  player_answer = float(input(f'Solve the equation: {equation} for x: '))

  if player_answer == (c - b) / a:
    print("That's correct!")
  else:
    print("This is mistake. Try it again.")

algebra_practice_game()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def projectile_game():
  wall_height = np.random.randint(5, 20)
  wall_position = np.random.randint(5, 15)

  plt.plot([wall_position, wall_position], [0, wall_height], color="red", linestyle="--", label="Wall")

  a= float(input("Enter the value of a: "))
  b= float(input("Enter the value of b: "))
  c= float(input("Enter the value of c: "))

  x = np.linspace(0, 20, 100)
  y = a* (x-b)**2 +c
  plt.plot(x, y, label="Projectile Path")

  plt.title("Projectile Game")
  plt.xlabel("X-axis")
  plt.ylabel("Y-axis")
  plt.legend()
  plt.show()
projectile_game()
```
